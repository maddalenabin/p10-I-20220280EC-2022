## For getting the qmap

```python
def get_qmap():
    andir = '/asap3/petra3/gpfs/p10/2021/data/11009212/processed/post-analysis/'
    
    qmap = np.load(andir+'geom/waxs-qmap-2d.npy')
    npzfile = np.load(andir+'geom/waxs-qmap-1d.npz')
    q = npzfile['q']
    idx = npzfile['idx']
    
    return qmap, q, idx
```

```python
def read_waxs_series(sample, first_n=10, last_n=10, norm=True, rundir='/asap3/petra3/gpfs/p10/2021/data/11009212/raw/'):

    series = sorted(glob(rundir + f'{sample}_*'))

    n = first_n + last_n
    qmap, q, idx = get_qmap()
    
    fig, ax = plt.subplots()
    colors = plt.cm.jet(np.linspace(1,0,len(series)))
    
    for i, counter in enumerate(series):
        frames = sorted(glob(counter+'/p300/*.cbf'))
        nframes = len(frames)
        #print(counter, ': ', nframes)

        image_in = np.empty(shape=(qmap.shape[0], qmap.shape[1], first_n))
        image_fin = np.empty(shape=(qmap.shape[0], qmap.shape[1], last_n))

        for k in range(first_n):#len(iruns)): range(start, stop[, step])
            #print('\t', k)
            # shape of image:  0=initial / 1=final, qmap.shape=detector (2d), number of frames read
            image_in[:,:,k] = load_cbf(frames[k])

        last_i = 0
        for k in range(nframes-last_n, nframes):
            #print('\t', k, last_i)
            image_fin[:,:,last_i] = load_cbf(frames[k])
            last_i += 1

        Iqin_avg = np.average(image_in, axis=-1)
        Iqfin_avg = np.average(image_fin, axis=-1)

        Iqin_flat = np.array(Iqin_avg).flatten()
        Iqfin_flat = np.array(Iqfin_avg).flatten()

        Iqin_flat_sort = Iqin_flat[idx]
        Iqfin_flat_sort = Iqfin_flat[idx]

        Iq_in = []
        Iq_fin = []

        for k in range(image_in.shape[0]):
            Iq_in.append(np.average(Iqin_flat_sort[k*image_in.shape[1]:(k+1)*image_in.shape[1]]))
            Iq_fin.append(np.average(Iqfin_flat_sort[k*image_fin.shape[1]:(k+1)*image_fin.shape[1]]))

        if norm is True:
            ax.plot(q, Iq_in/max(Iq_in[:160]), c=colors[i], ls='--', lw=1)
            ax.plot(q, Iq_fin/max(Iq_fin[:160]), c=colors[i], ls='-', lw=1, label=f'{counter[-3:]}')

        if norm is False:
            ax.plot(q, Iq_in, c=colors[i], ls='--', lw=1)
            ax.plot(q, Iq_fin, c=colors[i], ls='-', lw=1, label=f'{counter[-3:]}')

    Ih_ref = iceIh_peaks()
    for ice in Ih_ref: ax.axvline(x=ice, lw=1, alpha=.5, c='gray')

    ax.legend(loc='upper right', frameon=False, labelspacing=.0, handlelength=.7)
    #ax.set_ylim(0.3,1.1)
    ax.set_xlim(q[0],q[-1])
    ax.set_title(f'{sample}')
    ax.set_ylabel('I(q)', fontsize=14)
    ax.set_xlabel('q [$\AA^{-1}$]', fontsize=14)
    #ax.grid()
    plt.show()
```
